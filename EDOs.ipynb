```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def y_dot(t, y):
    return y
def euler(y0, a, b, n):
    ys = np.ndarray((n))
    h = (b-a)/n 
    ys[0] = y0
    for i in range(1, n):
        t=i*h
        ys[i] = ys[i-1] + h*y_dot(t, ys[i-1])
    return ys
```

```python
points = 10000
N = 20
solutions = np.ndarray((N, points))
for i in range(N):
    solutions[i] = euler(i, 0, 10, points)
```

```python
plt.figure(1, (16,8))
plt.grid()
for i in range(N):
    plt.plot(solutions[i])
```

```python
def runge_kutte(y0, a, b, n):
    ys = np.ndarray((n))
    h = (b-a)/n 
    ys[0] = y0
    for i in range(1, n):
        t=i*h
        ys[i] = ys[i-1] + (h/2)*(y_dot(t, ys[i-1]) + y_dot((i+1)*h, solutions[y0][i])) #only working for integers
    return ys
```

```python
N = 20
rk_solutions = np.ndarray((N, points))
for i in range(N):
    rk_solutions[i] = runge_kutte(i, 0, 10, points)
```

```python
plt.figure(1, (16,8))
plt.grid()
for i in range(N):
    plt.plot(rk_solutions[i])
```
